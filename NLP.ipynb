```python
import nltk
import re
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer

corpus = ['The cats were running faster than the dogs', 'Data Science involves analyzing large data sets']
```

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
stop_words=set(stopwords.words('english'))
```

```python
def preprocess_text(txt):
    txt=re.sub('[^a-zA-Z ]','',txt.lower())
    tokens = word_tokenize(txt)
    processed_tokens = [lemmatizer.lemmatize(wd) for wd in tokens if wd not in stop_words]
    return " ".join(processed_tokens)
```

```python
cleaned_corpus = [preprocess_text(doc) for doc in corpus]
vectorizer = CountVectorizer()
bow_matrix = vectorizer.fit_transform(cleaned_corpus)
print(cleaned_corpus)
print(vectorizer.get_feature_names_out())
print(bow_matrix.toarray())
```
